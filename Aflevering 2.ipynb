### Udvælge et tilfældigt tegn i en streng

d) 
random.randint(2, 5) genererer et tilfældigt heltal mellem 2 og 5, inklusive både 2 og 5.

```python
import random

def udvalgtTegn(streng):
    indeks = random.randint(0, len(streng)-1)  # Vi sætter variabel indeks, til at være et tilfældigt tal mellem 0 og indeks længde -1
    return streng[indeks]  # her returnere vi det indeks som er blevet udvalg tidligere, og det er den information som står på indeks plads. (et bogstav)
```

```python
udvalgtTegn('samtidig')
```

```python
import random

def udvalgtTegn(streng):
    return random.choice(streng) #returner et tilfældigt tegn i strengen
```

```python
udvalgtTegn('strawberry')
```

```python
udvalgtTegn(1234)
```

Fejlen i ovenstående kald af funktionen udvalgtTegn med parameteren 1234 er, at vi forsøger at returnere en streng, og tallene som er skrevet inde som parameter er ikke en streng men en int.
fejlen kommer fordi vi giver en int til en metode (random.choice()), der kræver en streng eller en liste

### Opgave 2

### Kryptering

```python
import random

def kryptere(streng):
    krypteret = ''  #Vi opretter en tom streng i en variabel som vi kalder krypteret
    alfabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZÆØÅabcdefghijklmnopqrstuvwxyzæøå'
    for tegn in streng: #i denne kryptering, tilfører vi kun 1 bogstav tilfældigt udvalgt pr. tegn i vores streng.
        krypteret += tegn #til (det pågældende) tegn til listen
        krypteret += random.choice(alfabet)  #når vi har tilføjet et tegn fra strengen, udvælger vi et tilfældigt bogstav fra vores streng alfabet.
    return krypteret
```

```python
kryptere('jordbær')
```

### Udvidet kryptering

```python
import random

def kryptere(streng, antal):
    krypteret = ''
    alfabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZÆØÅabcdefghijklmnopqrstuvwxyzæøå'
    for tegn in streng:
        krypteret += tegn
        for i in range(0, antal): #den nye kryptering skal være i stand til at indsætte et specifikt udvalgt antal tegn mellem hvert bogstav i vores streng, range her tager dog også højde
                                  #for tallet 0, hvis man ønsker dette
            krypteret += random.choice(alfabet)
    return krypteret
```

```python
kryptere('moren', -4)
```
