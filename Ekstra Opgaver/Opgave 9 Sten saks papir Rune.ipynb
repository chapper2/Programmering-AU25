```python
def spilRunde():
    brugerValg = bestemBrugerValg()
    computerValg = bestemComputerValg()
    udskrivValg(brugerValg, computerValg)
    resultat = bestemResultat(brugerValg, computerValg)
    udskrivResultat(resultat)
```

```python
import random

brugerSejre = 0
computerSejre = 0

def spilRunde():
    global brugerSejre, computerSejre
    brugerValg = bestemBrugerValg()
    computerValg = bestemComputerValg()
    udskrivValg(brugerValg, computerValg)
    resultat, b, c = bestemResultat(brugerValg, computerValg)
    brugerSejre += b
    computerSejre += c
    udskrivResultat(resultat, brugerSejre, computerSejre)
    bestemPrøvIgen()


def bestemBrugerValg():
    # Returnerer brugerens valg mellem sten, saks og papir
    valg = input("Vælg sten, saks eller papir: ").lower()
    while valg not in ['sten', 'saks', 'papir']:
        valg = input("Ugyldigt valg, prøv igen ved at skrive sten, saks eller papir: ").lower()
    return valg

def bestemComputerValg():
    # returnerer computerens valg - googl 'w3 random python' for at få hjælp til en passende metode ved random
    vilkårlig = random.choice(['sten', 'saks', 'papir'])
    return vilkårlig

def udskrivValg(brugerValg, computerValg):
    print(f"Du valgte {brugerValg}")
    print(f"Computer valgte {computerValg}")
    
def bestemResultat(brugerValg, computerValg):
    if brugerValg == computerValg:
        return "uafgjort", 0, 0
    elif (brugerValg == 'sten' and computerValg == 'saks') or \
         (brugerValg == 'saks' and computerValg == 'papir') or \
         (brugerValg == 'papir' and computerValg == 'sten'):
        return "Du vinder", 1, 0
    else:
        return "Computer vinder", 0, 1

def udskrivResultat(resultat, brugerSejre, computerSejre):
    print(resultat)
    print(f"Stilling: Du {brugerSejre} - Computer {computerSejre}")

def bestemPrøvIgen():
    prøvIgen = input("Vil du spille igen? (ja/nej): ").lower()
    if prøvIgen == 'ja':
        spilRunde()
    else:
        print("Tak for spillet!")
```
