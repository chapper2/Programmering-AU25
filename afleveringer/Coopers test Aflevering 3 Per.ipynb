### Python-kode med funktioner

```python
def kondital(meter):
    return (meter - 504.9)/44.73

def gennemsnitKondital(antalElever):
    sumKondital = 0
    for _ in range(antalElever): # for hver elev
        antalMeter = int(input('Indtast antal meter: ')) # antallet af meter løbet indlæses
        sumKondital += kondital(antalMeter)  # elevens kondital beregnes og lægges til summen af kondital
    return round(sumKondital/antalElever, 1)  # returnerer gennemsnittet med 1 decimal

def vurderKondital(snit):
    if snit < 37:
        return 'meget lavt'
    elif snit < 42:
        return 'lavt'
    elif snit < 51:
        return 'middel'
    elif snit < 56:
        return 'højt'
    else:
        return 'meget højt'   
```

```python
antal = int(input('Hvor mange elever har gennemført Cooperstesten? '))
snit = gennemsnitKondital(antal)

print('Klassens gennemsnitlige kondital er', snit)
print('Gennemsnittet vurderes til at være', vurderKondital(snit))
```
